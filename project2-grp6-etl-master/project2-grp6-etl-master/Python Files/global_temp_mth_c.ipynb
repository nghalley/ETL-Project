```python
import datapackage
import pandas as pd
```

```python
data_url = 'https://datahub.io/core/global-temp/datapackage.json'
```

```python
#load data paclage into storage
package = datapackage.Package(data_url)
```

```python
#load tabular data
resources = package.resources
for resource in resources:
    if resource.tabular:
        global_temp = pd.read_csv(resource.descriptor['path'])
        print(global_temp)
```

```python
gbl_temp_df = pd.DataFrame(global_temp)
print(gbl_temp_df)
```

```python
gbl_temp_df.to_csv('gbl_temp.csv')
```
